```python
import lyricsgenius
import re
art = "Playboi Carti"
key = "DeX6Z2v8AizIHuqVxkJ9GiOF04xyaZcnGNrXU_JD45yjKldZgwF-z3IUkv0U-dBI"
genius = lyricsgenius.Genius(key)
artist = genius.search_artist(art, max_songs=10)
artistsongslyrics = [x.lyrics for x in artist.songs]
cleanedlyrics  = []
alladlibs = []
for x in artistsongslyrics:
   x = str(x)
   adlibs = re.findall(r"\([^()]*\)", x)
   q = x
   q = q.replace("[" , "(")
   q = q.replace("]" , ")")
   q = re.sub(r" ?\([^)]+\)", "", q)
   q = q.split("\n")
   q = [a for a in q if a is not None and len(a) > 0]
   for z in list(set(q)):
      cleanedlyrics.append(z)
   for j in list(set(adlibs)):
     alladlibs.append(j)

from textgenrnn import textgenrnn
textgen = textgenrnn()
trainedmodel = textgen.train_on_texts(cleanedlyrics, num_epochs=5,gen_epoch = 5)
textgen.generate_to_file("carti.txt", n=5)
```
